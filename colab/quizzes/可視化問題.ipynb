## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
data1 = pd.read_csv('https://raw.githubusercontent.com/yutarofukushima-tech/Basic-python/master/colab/quizzes/matplotlib_quiz/hist_data.csv')
plt.hist(data1["aデータ"], bins=30, alpha=0.5, label='aデータ')
plt.hist(data1["bデータ"], bins=30, alpha=0.5, label='bデータ')
plt.title('aデータとbデータの温度についての度数分布')
plt.xlabel('温度')
plt.ylabel('度数')
plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
data2 = pd.read_csv('https://raw.githubusercontent.com/yutarofukushima-tech/Basic-python/master/colab/quizzes/matplotlib_quiz/scatter_data.csv')
x = data2['length (cm)']
y = data2['width (cm)']
color = data2['class']
plt.xlabel('length(cm)')
plt.ylabel('width(cm)')
plt.title('class0～2のlengthとwidthについての散布図')
plt.grid()
plt.scatter(x,y,c=color)
plt.legend
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(1).png

問3(2).png

```python
data3 = pd.read_csv('https://raw.githubusercontent.com/yutarofukushima-tech/Basic-python/master/colab/quizzes/matplotlib_quiz/line_data.csv')
x = data3['期間']
y = data3['件数']
z = data3['費用']
plt.plot(x, y, marker = 'o')
plt.xlabel('期間')
plt.xticks(rotation=45)
plt.ylabel('件数')
plt.grid(True)
plt.title('期間ごとの件数の推移')
plt.show()
```

```python
y = data3['費用']
plt.bar(x,y)
plt.grid()
plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.xticks(rotation=45)
plt.title('期間ごとの費用の推移')
plt.show()
```
